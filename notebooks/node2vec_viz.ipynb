```python
import hdbscan
import hvplot.pandas
import pandas as pd
import umap
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

import graphml.constants
```

```python
constants = graphml.constants.Constants()
```

```python
file = constants.DATA.joinpath('embed_cites.csv.gz')
data = pd.read_csv(file)
```

```python
data
```

```python
reducer = Pipeline([
    ('scaler', StandardScaler()),
    ('umap', umap.UMAP(n_neighbors=10, metric='cosine', n_components=2))
])

coords = reducer.fit_transform(data.iloc[:, 3:])

data = pd.concat(
    [
        data,
        pd.DataFrame(coords, columns=['embedX', 'embedY'])
    ],
    axis=1
)
```

```python
reducer = Pipeline([
    ('scaler', StandardScaler()),
    ('umap', umap.UMAP(n_neighbors=10, metric='cosine', n_components=10))
])

coords = reducer.fit_transform(data.iloc[:, 3:-2])

data = pd.concat(
    [
        data,
        pd.DataFrame(coords, columns=[f'umap_{i}' for i in range(10)])
    ],
    axis=1
)
```

```python
data.iloc[:, -12:]
```

```python
scaler = StandardScaler()
embeds = scaler.fit_transform(data.iloc[:, -10:])

cluster = hdbscan.HDBSCAN(min_cluster_size=40)
cluster.fit(embeds)

data['cluster_labels'] = cluster.labels_
```

```python
data['cluster_labels'].value_counts()
```

```python
sample = data.sample(frac=.1)
sample.head()
```

```python
sample.hvplot.scatter(
    x='embedX',
    y='embedY',
    by='category',
    title=f"Embedded Papers by Category",
    height=600,
    width=800,
    hover_cols=["title", "category"],
    legend=False,
    alpha=0.5
)
```

```python
sample.hvplot.scatter(
    x='embedX',
    y='embedY',
    by='cluster_labels',
    title=f"Embedded Papers by Cluster",
    height=600,
    width=800,
    hover_cols=["title", "category", "cluster_labels"],
    legend=False,
    alpha=0.5
)
```

```python
data.hvplot.scatter(
    x='embedX',
    y='embedY',
    by='cluster_labels',
    title=f"Embedded Papers by Cluster",
    height=600,
    width=800,
    hover_cols=["title", "category", "cluster_labels"],
    legend=False,
    alpha=0.5,
    datashade=True
)
```

```python
data.to_csv(
    constants.DATA.joinpath('embed_cites_umap.csv.gz'),
    compression='gzip',
    index=False,
)
```
